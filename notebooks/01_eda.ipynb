## Feature Exploration

```python
# Distribution of transaction amount
plt.figure(figsize=(8,4))
sns.histplot(df['Amount'], bins=50, kde=True)
plt.title('Transaction Amount Distribution')
plt.savefig('../reports/eda/amount_distribution.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
import os

# Ensure the directory exists
os.makedirs('../reports/eda/violin_plots/', exist_ok=True)

# List of features to plot
features = [f'V{i}' for i in range(1, 29)] + ['Amount']

# Set up the subplot grid
n_cols = 3
n_rows = -(-len(features) // n_cols)  # Ceiling division
fig, axes = plt.subplots(n_rows, n_cols, figsize=(18, n_rows * 4))

# Flatten axes array for easy iteration
axes = axes.flatten()

# Plot each feature
for idx, col in enumerate(features):
    sns.violinplot(x='Class', y=col, data=df, ax=axes[idx], inner='quartile')
    axes[idx].set_title(f'{col} vs Class')
    axes[idx].set_xlabel('')
    axes[idx].set_ylabel('')
    
    # --- Save each subplot individually ---
    extent = axes[idx].get_window_extent().transformed(fig.dpi_scale_trans.inverted())
    fig.savefig(f'../reports/eda/violin_plots/{col}_vs_class.png',
                bbox_inches=extent.expanded(1.2, 1.3), dpi=300)

# Remove any unused subplots
for j in range(idx + 1, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```

```python
# Create a new column for the log-transformed Amount
df['LogAmount'] = np.log1p(df['Amount'])  # log1p = log(1 + x) handles zeros safely

# Plot the violin plot
plt.figure(figsize=(8, 4))
sns.violinplot(x='Class', y='LogAmount', data=df, inner='quartile')
plt.title('Log-Transformed Amount vs Class')
plt.xlabel('Class')
plt.ylabel('log(1 + Amount)')

# plt.savefig('../reports/eda/log_transformed_amount_vs_class.png', dpi=300, bbox_inches='tight')

plt.show()
```

## Correlation Analysis

```python
# Compute correlation matrix
corr_matrix = df.corr()

# Set up the heatmap
plt.figure(figsize=(16, 12))
sns.heatmap(corr_matrix, cmap='coolwarm', linewidths=0.5, annot=False)
plt.title('Correlation Matrix Heatmap')

# plt.savefig('../reports/eda/correlation_matrix_heatmap.png', dpi=300, bbox_inches='tight')

plt.show()
```

```python
# Calculate correlations with 'Class'
correlations = corr_matrix['Class'].drop('Class')  # exclude self-correlation

# Get top 10 features with strongest absolute correlation (positive or negative)
top_10_features = correlations.abs().sort_values(ascending=False).head(10).index

# Show the correlations (with sign)
top_10_corr = correlations.loc[top_10_features]
print(top_10_corr)
```

```python
plt.figure(figsize=(10, 6))
top_10_corr_sorted = top_10_corr.sort_values()

plt.figure(figsize=(10, 6))
bars = plt.barh(top_10_corr_sorted.index, top_10_corr_sorted.values, 
                color=['green' if x > 0 else 'red' for x in top_10_corr_sorted.values])
plt.title('Top 10 Features Correlated with Class')
plt.xlabel('Correlation coefficient')
plt.ylabel('Feature')
plt.grid(axis='x')
plt.tight_layout()

# Save figure
plt.savefig('../reports/eda/top_10_correlations_with_class.png', dpi=300, bbox_inches='tight')
plt.show()
```

## Time-Based Plots

```python
df['HourOfDay'] = ((df['Time'] // 3600) % 24).astype(int)

# Calculate fraud ratio by hour of day
hourly_fraud_ratio = df.groupby('HourOfDay')['Class'].mean()

plt.figure(figsize=(12, 5))
hourly_fraud_ratio.plot(kind='bar', color='purple')
plt.title('Fraud Ratio by Hour of Day')
plt.xlabel('Hour of Day (0â€“23)')
plt.ylabel('Fraction of Transactions that are Fraud')
plt.tight_layout()
plt.savefig('../reports/eda/fraud_ratio_by_hour_of_day.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
# Group by hour of day
hourly_stats = df.groupby('HourOfDay')['Class'].agg(
    Total_Transactions='count',
    Total_Frauds='sum'
)
hourly_stats['Fraud_Ratio'] = hourly_stats['Total_Frauds'] / hourly_stats['Total_Transactions']

# Plot
fig, ax1 = plt.subplots(figsize=(12, 6))

# Bar for total transactions
ax1.bar(hourly_stats.index, hourly_stats['Total_Transactions'], color='skyblue', label='Total Transactions')
ax1.set_ylabel('Total Transactions', color='skyblue')
ax1.set_xlabel('Hour of Day (0â€“23)')
ax1.tick_params(axis='y', labelcolor='skyblue')

# Line for fraud ratio
ax2 = ax1.twinx()
ax2.plot(hourly_stats.index, hourly_stats['Fraud_Ratio'], color='red', marker='o', label='Fraud Ratio')
ax2.set_ylabel('Fraud Ratio', color='red')
ax2.tick_params(axis='y', labelcolor='red')

plt.title('Transaction Volume and Fraud Ratio by Hour of Day')
fig.tight_layout()
plt.savefig('../reports/eda/transactions_fraud_ratio_by_hour_of_day.png', dpi=300, bbox_inches='tight')
plt.show()
```
